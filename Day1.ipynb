```python
left=[]
right=[]
with open("input1.txt","r") as f:
    for line in f:
        list = line.rstrip().split("   ",1)
        bisect.insort(left,int(list[0]))
        bisect.insort(right,int(list[1]))

f.close()
```

```python
length = 0
for i in range(len(left)):
    length+= abs(left[i]-right[i])
print (length)
```
