```python
def cmp(a,b):
    return (a>b)-(a<b)
```

```python
def checksafe1(list):
    safe = True
    sign0 = cmp(int(list[1])-int(list[0]),0)
    if sign0 == 0:
        return False
    for i in range(len(list)-1):
        sign = cmp(int(list[i+1])-int(list[i]),0)
        if sign0 != sign or abs(int(list[i+1])-int(list[i]))>3:
            safe = False
            break
    return safe
```

```python
safe=0
with open("input2.txt","r") as f:
    for line in f:
        list = line.rstrip().split(" ")
        safe+=checksafe1(list)
        ###print(list)
print(safe)
f.close()
```
